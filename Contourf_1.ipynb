```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

```python
def f(x,y):
    return  np.exp(x**5)*np.exp(y**5) + 2*x + 3*y
```

```python
a, b = plt.subplots()
b.set_xlabel('x', fontsize = 14)
b.set_ylabel('y', fontsize = 14)

n = 500
x = np.linspace(-2, 2, n)
y = np.linspace(-2, 2, n)

lv = [0.05,0.3,1,2,4,6]
cf = [0,0.05,0.3,1,2,4,6,np.exp(2)]

X,Y = np.meshgrid(x,y)
plt.contourf(X,Y,f(X,Y),levels=cf,alpha=0.75, cmap=plt.cm.hot)
C = plt.contour(X,Y,f(X,Y),levels=lv,color='black',lw=0.5, cmap=plt.cm.hot)
plt.clabel(C, inline = True ,fontsize = 10)
```
